```python
 titanic_data = sns.load_dataset('titanic')
```

```python
#Box plot for distribution of age with respect to each gender and survival status
plt.figure(figsize=(12, 8))
boxplot = sns.boxplot(x='sex', y='age', hue='survived', data= titanic_data, palette='Set2')
plt.title('Distribution of Age with Respect to Gender and Survival Status')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```
