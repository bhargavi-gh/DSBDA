```python
stud_data["reading score"]=stud_data["reading score"].replace(np.NaN,stud_data["reading score"].mean())
#replacing NaN values by mean
```

```python
stud_data
```

```python
stud_data["writing score"]=stud_data["writing score"].replace(np.NaN,stud_data["writing score"].median())
#replacing NaN values by median
```

```python
stud_data
```

```python
stud_data.isnull().sum() #checking if there are any NaN values in the dataframe
```

```python
stud_data["math score"]=stud_data["math score"].replace(np.NaN,statistics.mode(stud_data["math score"]))
#replacing NaN values by mode
```

```python
stud_data
```

```python
stud_data.isnull().sum()
```
